```python
import numpy as np
import matplotlib.pyplot as plt
from PyEMD import EMD

# 加载地震波数据
data = np.loadtxt('1.txt')

# 生成时间轴（假设数据是均匀采样的）
time = np.arange(len(data))

# 进行EMD分解
emd = EMD()
IMFs = emd(data)

# 绘制原始地震波数据
plt.figure(figsize=(10, 6))
plt.subplot(len(IMFs) + 1, 1, 1)
plt.plot(time, data, 'b')
plt.title('Original Seismic Wave')

# 绘制各个IMF
for i, IMF in enumerate(IMFs):
    plt.subplot(len(IMFs) + 1, 1, i + 2)
    plt.plot(time, IMF, 'r')
    plt.title('IMF {}'.format(i + 1))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# 加载地震波数据
data = np.loadtxt('1.txt')

# 小波包分解
level = 5  # 小波包分解的层数
wavelet = 'db4'  # 使用的小波基函数

coeffs = pywt.wavedec(data, wavelet, level=level)

# 绘制小波包分解系数
plt.figure(figsize=(10, 6))
plt.subplot(level + 2, 1, 1)
plt.plot(data, 'b')
plt.title('Original Seismic Wave')

for i in range(level + 1):
    plt.subplot(level + 2, 1, i + 2)
    plt.plot(coeffs[i], 'r')
    plt.title('Level {}'.format(i))

plt.tight_layout()
plt.show()
```
